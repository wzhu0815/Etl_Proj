```python
datas.to_csv("Output/datas.csv",index=False)
```

```python
# iterating the columns 
for col in datas.columns: 
    print(col) 
```

```python
cancelled_df=datas.loc[data1["CANCELLED"] == 1]
cancelled_df.head()
```

```python
cancelled2_df = cancelled_df.drop(['CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY','DIVERTED','ACTUAL_ELAPSED_TIME','AIR_TIME','Unnamed: 27','ARR_TIME','ARR_DELAY','DEP_TIME','DEP_DELAY','TAXI_OUT','WHEELS_OFF','WHEELS_ON','TAXI_IN'], axis=1)
cancelled2_df.head()
```

```python
#cancelled2_df.to_csv("Output/cancelled2.csv", index=False, header=True)
```

```python
#start process to merge cancel codes
cancelcode="Resources/cancellation_codes.csv" 
# Preview the first 5 lines of the loaded data 
cancelcd_df = pd.read_csv(cancelcode)
cancelcd_df.head()
```

```python
#rename column for join
cancelcd_df.rename(columns = {"CANCELLATION_REASON": "CANCELLATION_CODE"},  
          inplace = True) 
cancelcd_df.head()
```

```python
#join dfs on cancellation code
cancel2_df = pd.merge(cancelled2_df, cancelcd_df, on="CANCELLATION_CODE", how="outer")
cancel2_df.head()
```

```python
#split date column onto 3 col; m/d/y
cancel2_df['FL_DATE']= pd.to_datetime(cancel2_df['FL_DATE'],format='%Y-%m-%d')
cancel2_df['FL_DATE']
```

```python
#split date column;mm/day(0=sumday)/year may be useful in analysis
cancel2_df['year']= cancel2_df['FL_DATE'].dt.year
cancel2_df['month']= cancel2_df['FL_DATE'].dt.month
cancel2_df['day']= cancel2_df['FL_DATE'].dt.dayofweek
```

```python
cancel3_df=cancel2_df.append(['year','month', 'day'])
cancel3_df.head()
```

```python
#cancel3_df.drop(column=['0'], inplace=True)
for col in cancel3_df.columns: 
    print(col) 
```

```python
#mongo accepts keys instring format also needed to drop extra indx column
cancel3_df=cancel3_df.applymap(str)
```

```python
cancel4_df=cancel3_df.drop(cancel3_df.columns[0], axis = 1) 
cancel4_df.head()
```

```python
cancel4_df.to_csv("Output/cancel4.csv", index=False, header=True)
```
